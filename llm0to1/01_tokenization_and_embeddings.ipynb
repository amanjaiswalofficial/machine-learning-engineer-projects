```python
from transformers import AutoTokenizer
model_name = "bert-base-uncased"  # Pre-trained BERT tokenizer
model_name = 'gpt2'

text = "Tokenization is the first step in understanding language models"
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokens = tokenizer.tokenize(text)
token_ids = tokenizer.convert_tokens_to_ids(tokens)
decoded_text = tokenizer.decode(token_ids)

print(f"Original: {text},\n Tokens: {tokens},\n Token IDs: {token_ids},\n Decoded Text: {decoded_text}")
```

## Embeddings

1.  Create a single vector for each sentence
2. A similarity score close to 1 means the sentences are very similar in meaning, while a score closer to 0 means they are not.
3. Word and sentence embeddings allow machines to understand the semantic relationships between words and sentences:
4. Words like "king" and "queen" have embeddings that are closer in vector space.
5. Similarly, sentences like "I enjoy pizza" and "Pizza is my favorite food" will have similar embeddings.
6. Useful in: Finding similar documents, Classification, Question/Answers
7. Teaches how text is represented in models

```python
from transformers import AutoTokenizer, AutoModel
import torch
from sklearn.metrics.pairwise import cosine_similarity

model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

text = ["Machine learning is amazing!",
        "Apples and bananas are differen"]
inputs = tokenizer(text, padding=True, truncation=True, return_tensors="pt")

# Get word embeddings
with torch.no_grad():
  outputs = model(**inputs)
  # The last hidden state contains embeddings for each token
  embeddings = outputs.last_hidden_state

# Calculate sentence embeddings (mean pooling, same as averaging)
sentence_embeddings = torch.mean(embeddings, dim=1)

# Calculuate cosine similarity bewteen sentences
similarity = \
  cosine_similarity(sentence_embeddings[0].unsqueeze(0),
                    sentence_embeddings[1].unsqueeze(0))

print(f"Sentence 1 Shape: {sentence_embeddings[0].shape}")
print(f"Sentence 1 Shape: {sentence_embeddings[0].shape}")
print(f"Cosine similarity b/w senences: {similarity[0][0]:.4f}")
```
