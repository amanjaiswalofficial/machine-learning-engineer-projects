```python
import pandas as pd
import numpy as np

CHUNKSIZE = 500_000
EVENT_TYPES = ['view']  # Keep only 'view' events for now
OUTPUT_EVENTS_FILE = "filtered_events.feather"

# ---------- STEP 1: Load events.csv in chunks ----------
event_chunks = []
reader = pd.read_csv("ecommerce-dataset/events.csv", chunksize=CHUNKSIZE)

for i, chunk in enumerate(reader):
    # Filter only view events
    chunk = chunk[chunk['event'].isin(EVENT_TYPES)]

    # Downcast columns to save RAM
    chunk['timestamp'] = pd.to_numeric(chunk['timestamp'], downcast='integer')
    chunk['visitorid'] = pd.to_numeric(chunk['visitorid'], downcast='integer')
    chunk['itemid'] = pd.to_numeric(chunk['itemid'], downcast='integer')

    event_chunks.append(chunk)
    print(f"Processed chunk {i+1}")

# Concatenate all filtered chunks
events_df = pd.concat(event_chunks, ignore_index=True)

# Save to feather to avoid repeating chunking
events_df.to_feather(OUTPUT_EVENTS_FILE)
print(f"Saved filtered events to {OUTPUT_EVENTS_FILE}")
```

```python
# Function to process item_properties file (part1 or part2)
def process_item_properties(file):
    print(f"Processing: {file}")
    df = pd.read_csv(file)

    # Downcast to reduce memory
    df['itemid'] = pd.to_numeric(df['itemid'], downcast='integer')
    df['timestamp'] = pd.to_numeric(df['timestamp'], downcast='integer')

    return df

# Load both parts
part1 = process_item_properties("ecommerce-dataset/item_properties_part1.csv")
part2 = process_item_properties("ecommerce-dataset/item_properties_part2.csv")

# Combine
item_properties = pd.concat([part1, part2], ignore_index=True)
del part1, part2  # Free RAM

# Optional: Filter only items from events_df to reduce size
valid_item_ids = events_df['itemid'].unique()
item_properties = item_properties[item_properties['itemid'].isin(valid_item_ids)]

# Save reduced properties
item_properties.to_feather("filtered_item_properties.feather")
print("Filtered item_properties saved.")
```

## 02. Create clean item-level feature table

```python
import pandas as pd

filtered_events = pd.read_feather("filtered_events.feather")
viewed_item_ids = filtered_events['itemid'].unique()

filtered_item_properties = pd.read_feather("filtered_item_properties.feather")

filtered_item_properties = filtered_item_properties.drop(columns=["timestamp"])

filtered_item_properties = filtered_item_properties[
    filtered_item_properties['itemid'].isin(viewed_item_ids)
]

pivoted = filtered_item_properties.pivot_table(
    index="itemid",
    columns="property",
    values="value",
    aggfunc="first"  # Use the first value if duplicates exist
).reset_index()

pivoted
```
